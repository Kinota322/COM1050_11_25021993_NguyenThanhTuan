```python
#6

def prime_num(num):
    if num <= 1:
        return False
    for i in range(2, int((num ** (1/ 2)) + 1)):
        if num % i == 0:
            return False
    return True

def solve(s1, s2):
    res = 0
    for i in range(s1, s2 + 1):
        if prime_num(i):
            res += i
    return res

while(True):
    a, b = map(int, input().split())
    print(solve(a, b))
```

```python
#7
def prime_num(num):
    if num <= 1:
        return False
    for i in range(2, int((num ** (1/ 2)) + 1)):
        if num % i == 0:
            return False
    return True

n = int(input())
res = (prime_num(n) == True) * n
for i in range(1, int((n ** (1/2)) + 1)):
    if n % i == 0:
        res = max(res, (n / i) * prime_num(int(n / i)))
        res = max(res, i * prime_num(i))
print(res) 
```

```python
#8
def rev(num):
    res = 0
    while num:
        res *= 10
        res += num % 10
        num //= 10
    return res

n = int(input())
cnt = 0
while n != rev(n):
    cnt += 1
    n = n + rev(n)
print(cnt, n)
```

```python
#10
n = int(input())
def len_collatz(num):
    cnt = 1
    while num != 1:
        cnt += 1
        if num & 1:
            num = (3 * num) + 1
        else:
            num = num // 2
    return cnt

num_res = 1
res = 1
for i in range(1, n + 1):
    temp = len_collatz(i)
    if temp > res:
        num_res = i
        res = len_collatz(i)

print(num_res, res)
```

```python
#13
def tong_uoc(num):
    res = 1
    for i in range(2, int((num ** (1/ 2)) + 1)):
        if num % i == 0:
            res += i
            if (num // i) != i:
                res += num // i
    return res
a, b = map(int, input().split())

if tong_uoc(a) == b and a == tong_uoc(b):
    print('true')
else:
    print('false')
```
