```python
#9
inp1 = input()
inp2 = input()

res = {}

def solve(string):
    global res
    dic = {"'": 0, ":": 0}
    key = ''
    value = 0
    for index, char in enumerate(string):
        if char == "'" and dic["'"] != 0:
            key = string[(dic["'"] + 1): index]
            dic["'"] = 0
        elif char == "'" and dic["'"] == 0:
            dic["'"] = index

        if char == ":":
            dic[":"] = index

        if (char == "," or char == "}"):
            value = string[(dic[":"] + 2): index]
            if key != '' and res.get(key) == None:
                res[key] = 0
            if key != '':
                res[key] += int(value)
                key = ''

        # print(index, key, value)

solve(inp1)
solve(inp2)

print(res)
```

```python
#12
inp = list(map(int, input().split()))

res_times = 0
res = inp[0]

cnt = {}

for i in inp:
    if cnt.get(i) == None:
        cnt[i] = 0
    
    cnt[i] += 1

    if cnt[i] > res_times:
        res_times = cnt[i]
        res = i
    elif cnt[i] == res_times:
        res = min(res, i)
    
    # print(res)

print(res)
```

```python
#14
try:
    line1 = list(map(int,input().split())) 
except EOFError:
    line1 = []

try:
    line2 = list(map(int,input().split()))
except EOFError:
    line2 = []

res = []
cnt = {}

for i in line2:
    cnt[i] = 1
for i in line1:
    if cnt.get(i) != None and cnt[i] == 1:
        res.append(i)
        cnt[i] = 0
print(res)
```

```python
#18
n, m, k = map(int, input().split())

matrix = [[0] * m] * n

for i in range(n):
    matrix[i] = list(map(int, input().split()))

sum_row = 0
sum_col = 0

for i in range(n):
    sum_col += matrix[i][k - 1]
for i in range(n):
    sum_row += matrix[k - 1][i]

print("sum col k", sum_col)
print("sum row k", sum_row)
```
