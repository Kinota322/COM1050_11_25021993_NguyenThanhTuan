```python
#3
num = int(input())

def len_bit(n):
    res = 0
    while n:
        res += 1
        n = n >> 1
    return res

def check(n):
    if n == (1 << (len_bit(n) - 1)):
        return True
    else:
        return False

print(check(num))
```
