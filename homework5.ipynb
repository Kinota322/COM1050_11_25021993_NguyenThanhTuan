```python
#1
def Max_Of_Two(num1, num2):
    if num1 > num2:
        return num1
    else:
        return num2

a, b = map(int, input())

print(Max_Of_Two(a, b))
```

```python
#2
def swap(num1, num2):
    temp_tuple = (num2, num1)
    return temp_tuple

a, b = map(int, input().split())

print(swap(a, b)[0], swap(a, b)[1])
```

```python
#3
def prime_num(num):
    if num <= 1:
        return False
    for i in range(2, int((num ** (1/ 2)) + 1)):
        if num % i == 0:
            return False
    return True

n = int(input())

print(prime_num(n))
```

```python
#4
def perfect_num(num):
    temp = 1
    for i in range(2, int((num ** (1/ 2)) + 1)):
        if num % i == 0:
            temp += i
            if num // i != i:
                temp += num // i
    if num == temp:
        return True
    else:
        return False
n = int(input())

print(perfect_num(n))
    
```

```python
#5
l = list(map(int, input().split()))
k = int(input())
res = -1
for i in range(1, len(l) + 1):
    if l[i - 1] == k:
        res = i
        break
print(res)
```

```python
#6
def giaithua(num):
    temp = 1
    for i in range(2, num + 1):
        temp *= i
    return temp

n = int(input())

print(giaithua(n))
```

```python
#7
def cal(List):
    res = 0
    if List[1] == '+':
        res = (int(List[0]) + int(List[2]))
    if List[1] == '-':
        res = (int(List[0]) - int(List[2]))
    if List[1] == '/':
        res = (int(List[0]) / int(List[2]))
    if List[1] == '*':
        res = (int(List[0]) * int(List[2]))
    return res
l = input().split()
print(f'{cal(l):.2f}')
```

```python
#8
def hamming_distance(num1, num2):
    cnt = 0
    while num1 or num2:
        cnt += (num1 & 1) ^ (num2 & 1)
        num1 >>= 1
        num2 >>= 1
    return cnt

a, b = map(int,input().split())

print(hamming_distance(a, b))
```

```python
#9
def tong_chu_so(num):
    res = 0
    while num:
        res += num % 10
        num //= 10
    return res

n = int(input())

print(tong_chu_so(n))
```

```python
#10
def check(str1, str2):
    func = {}
    for i in range(0, len(str1)):
        if func.get(str1[i]) == None:
            func[str1[i]] = str2[i]
        elif func.get(str1[i]) != str2[i]:
            return False
    return True

a, b = input().split()

if check(a, b) and check(b, a):
    print('true')
else:
    print('false')
```
